```python
import matplotlib.pyplot as plt
import numpy as np
%natplotlib inline
#%matplotlib具體作用是當你調用matplotlib.pyplot的繪圖函數plot()進行繪圖的時候
# 或者生成一個figure畫布的時候，可以直接在你的python console裏面生成圖像
```

```python
y = np.random.randint(-5,5 ,size=20)
print(y)
plt.plot(y)
```

```python
fig, ax = plt.subplots()
ax.plot(y)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])#([0,0<起始,1<寬,1<高])
ax1=fig.add_axes([0.5,0,0.5,0.5])
ax1.plot(y)
ax.plot(y)
```

```python
fig = plt.figure()
ax = plt.axes()
ax.plot(y)
```

```python
plt.plot(y,"--xb")
```

```python
plt.plot(y,linestyle='solid',
lw=10,color=(1,0,0),marker='p'
,markersize=20,
markerfacecolor='#FFF000',markeredgecolor='#FFFFFF')#linestyle='solid'線的樣式,lw=10線的粗細
#color=(1,0,0)線的顏色,marker='p'點的符號,markersize=20點的大小,
# markerfacecolor='FFF000'點的顏色markeredgecolor='#FFFFFF'點)
```

```python
plt.rcParams['font.sans-serif']=['Microsoft JhengHei']#檢視中文['Microsoft JhengHei']字形
plt.rcParams['axes.unicode_minus'] = False
x = np.linspace(-2*np.pi, 2*np.pi, 300)
y = np.sin(x)
plt.plot(x,y)
plt.title("sine function")
plt.xlabel('角度')
plt.ylabel("sin(x)")
plt.xticks([-5,0,4,5,6])
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_xticks(np.arange(-1.5*np.pi,2*np.pi,np.pi))
ax.set_xticklabels(f'{v}pi' for v in np.arange(-1.5,2,1))
```

```python
x = np.linspace(-2*np.pi,2*np.pi, 300)
y = np.sin(x)
y2 = np.cos(x)
plt.plot(x,y,label="sin(x)")
plt.plot(x,y2,label="cos(x)")
plt.plot(x,x)
plt.title("三角函數")
plt.xlabel('角度')
plt.legend(loc="lower right",frameon=True,shadow=True,fancybox=False)
plt.xlim([-2,2])
plt.ylim([-1,1])
plt.axis([-2,2,-1,1])
```

```python
x = np.linspace(-2*np.pi,2*np.pi,30)
y = np.sin(x)
plt.plot(x,y,'--bo',markersize=10)
```

```python
plt.scatter(x,y,marker='o',s=np.abs(x)*10,c=y,alpha=0.5)
plt.colorbar()
```

```python
areas = ["北","中","南"]
rain = [600,300,100]
plt.bar(areas, rain)
```

```python
areas = ["北","中","南"]
years = [2019,2020,2021]
rain =np.array(([600,300,100],[300,200,200],[500,200,600]))

sumRain = np.zeros(rain.shape[1])
for i in range(rain.shape[0]):
    plt.bar(areas, rain[i,:],bottom=sumRain)#由rain[0,:]往上疊
    sumRain += rain[i,:]
plt.legend(years)
plt.savefig('bar.jpg')#生成.jpg檔
```
