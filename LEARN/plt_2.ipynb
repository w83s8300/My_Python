```python
data = np.random.randn(1000)
plt.hist(data)
```

```python
plt.hist(data, bins=30)
```

```python
plt.hist(data, bins=[-5, -1, 0, 1 ,5])
```

```python
plt.hist(data, density=True, cumulative=True)
```

```python
mean = [7,1]
cov = [[3,5],[1,0]]
plt.rcParams['font.sans-serif']=['Microsoft JhengHei']#檢視中文['Microsoft JhengHei']字形
x1, x2 = np.random.multivariate_normal(mean, cov, 10000).T
plt.hist2d(x1,x2, bins=100)
cd = plt.colorbar()
cd.set_label('個數')
```

等高線圖contour

```python
x = np.linspace(-np.pi,np.pi,100)
y = np.linspace(-np.pi,np.pi,100)
xx ,yy =np.meshgrid(x,y)
z = np.sin(xx)*20 + np.sin(yy)*10
plt.contour(xx ,yy ,z)
plt.colorbar()
```

```python
plt.contourf(xx,yy,z,10,cmap="RdGy")
plt.colorbar
```

```python
grid = plt.GridSpec(3,3,hspace=0.3,wspace=0.1)
plt.subplot(grid[0,:-1])
plt.subplot(grid[:-1,-1])
plt.subplot(grid[-1,1:])
plt.subplot(grid[1:,0])
plt.subplot(grid[1,1])
```

```python
mean = [1,1]
cov = [[1,1],[1,0]]
x, y = np.random.multivariate_normal(mean, cov ,3000).T

fig =plt.figure(figsize=(12,12))
grid = plt.GridSpec(4,4,hspace=.01,wspace=0.01)
ax_xy = plt.subplot(grid[:-1,1:])#xy的位子
ax_x = plt.subplot(grid[-1,1:],yticks=[])#x的位子
ax_y = plt.subplot(grid[:-1,0],xticks=[])#y的位子
ax_xy.grid(True)#xy的圖加底線
ax_y.grid(True)#xy的圖加底線
ax_x.grid(True)#xy的圖加底線
ax_xy.plot(x,y,'ok',markersize=3)
ax_x.hist(x,40)
ax_x.invert_yaxis()#x圖的y變相反
ax_y.hist(y,40,orientation='horizontal')#y圖的上下變左右
ax_y.invert_xaxis()#y圖的x變相反
```
