```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
%matplotlib inline
```

```python
fig = plt.figure(figsize=(12,8))
x = np.r_[np.random.normal(loc=(0,2),size=(1000,2)),np.random.normal(loc=(2,1),size=(500,2)),np.random.normal(loc=(2,3),size=(500,2))]
y = np.r_[np.zeros(1000),np.ones(500),np.ones(500)]
plt.scatter(x[:,0],x[:,1],c=y,s=5)
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=1/3.0)#將data.data和data.target產生資料拆分成訓練及測試二組。test_size=1/3.0
```

```python
svc=SVC()
svc.fit(x_train, y_train)#在fit是只(數據集)（X，y）上使用SVM模型
y_pred =svc.predict(x_test)
print(f'accuracy_score:{accuracy_score(y_test,y_pred)}')
```

```python
#畫圖
xmin, xmax = x[:,0].min()-1, x[:,0].max()+1#x的範圍
ymin, ymax = x[:,0].min()-1, x[:,0].max()+1#y的範圍
xx ,yy =np.meshgrid(np.linspace(xmin,xmax,100), np.linspace(ymin,ymax,100))#
yhat = svc.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)#
plt.figure(figsize=(12,8))
plt.pcolormesh(xx,yy,yhat)##畫線圖
#plt.scatter(xx,yy,c=yhat,s=3)#畫點圖


```

```python
N=1000#你的點
angle = np.random.rand(N)*np.pi*4
r1 = angle +np.pi
dataA=np.array([r1*2*np.cos(angle),r1*2*np.sin(angle)]).T +np.random.randn(N,2)


r2 = -angle -np.pi
dataB=np.array([r2*2*np.cos(angle),r2*2*np.sin(angle)]).T +np.random.randn(N,2)

plt.scatter(dataA[:,0],dataA[:,1],s=5)
plt.scatter(dataB[:,0],dataB[:,1],s=5)

x = np.r_[dataA,dataB]
y = np.r_[np.zeros(N),np.ones(N)]
```

```python
#畫圖
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
fig = plt.figure(figsize=(16,8))
lr =LinearRegression()
for d in range(2,3,1):
    poly = PolynomialFeatures(degree=d,include_bias=False)
    xd =poly.fit_transform(x)
    lr.fit(xd, y)
    y_pred = lr.predict(xd)
    plt.plot(x ,y_pred,label=f'{d}-degre')
plt.legend()
```

```python
svc = SVC(C=100)#C=100是只條變質
svc.fit(x,y)
y_pred = svc.predict(x)
print(f'accuracy_score:{accuracy_score(y,y_pred)}')

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=1/3.0)#將data.data和data.target產生資料拆分成訓練及測試二組。test_size=1/3.0

svc.fit(x_train,y_train)
y_pred = svc.predict(x_train)
print(f'accuracy_score:{accuracy_score(y_train,y_pred)}')
y_pred = svc.predict(x_test)
print(f'accuracy_score:{accuracy_score(y_test,y_pred)}')
```

```python
N=5000#你的點
angle = np.random.rand(N)*np.pi*6
r1 = angle +np.pi
dataA=np.array([r1*np.cos(angle),r1*np.sin(angle)]).T +np.random.randn(N,2)


r2 = -angle -np.pi
dataB=np.array([r2*np.cos(angle),r2*np.sin(angle)]).T +np.random.randn(N,2)

plt.scatter(dataA[:,0],dataA[:,1],s=3)
plt.scatter(dataB[:,0],dataB[:,1],s=3)

x = np.r_[dataA,dataB]
y = np.r_[np.zeros(N),np.ones(N)]
```

```python
mlp = MLPClassifier(verbose=2,max_iter=1000, hidden_layer_sizes=(100,100,50))#max_iter最大疊代次數 hidden_layer_sizes 隱藏城的深度
mlp.fit(x,y)#fit(x,y)要學習的資料
y_pred= mlp.predict(x)
print(f'accuracy_score:{accuracy_score(y,y_pred)}')
```
